```python
#https://www.youtube.com/watch?v=yV2BCFvXmC8
#Anhanguera
#Como importar arquivos .CSV para seu projeto Google Colab, Jupyter Notebook ou Spider

import pandas as pd

informacao_df = pd.read_csv('bdmed192021.csv') #importar arquivo bdmed192021.csv para COLAB

display(informacao_df.head(20))#mostra as 10 primeiras linhas
display(informacao_df.shape)#mostra quantidade de linha e colunas
display(informacao_df.describe())#mostra um resumo
print("="*90)

print('CALCULAR INDICADORES utilizando values_counts PATOLOGISTA/ANO')
produtividade_pato = informacao_df[['Patologista Alocado','Ano']].value_counts()
display(produtividade_pato)

print("H"*90)
print('CALCULAR INDICADORES utilizando values_counts EXAME/ANO')
produtividade_ano = informacao_df[['Exame','Ano']].value_counts()
display(produtividade_ano)

print("H"*90)
print('CALCULAR INDICADORES utilizando values_counts PRIORIDADE/PERIODO')
produtividade_mes = informacao_df[['Prioridade','Ano','Mês']].value_counts()
display(produtividade_mes.head(40))#head(40)) exibe 40 linhas

print("="*90)
print('CALCULAR INDICADORES utilizando groupby mean média')
#sum = soma o total por exame ou mean média
faturamento_exame = informacao_df[['Exame','Espera']].groupby('Exame').mean()
display(faturamento_exame)
faturamento_blocos = informacao_df[['Exame','Quantidade de Blocos']].groupby('Exame').sum()
display(faturamento_blocos)
print('-'*90)
print('CALCULAR INDICADORES utilizando values_counts')
produtividade_exame = informacao_df['Exame'].value_counts()
display(produtividade_exame)
print("="*90)
faturamento_exame = informacao_df[['Patologista Alocado','Exame','Espera']].groupby(['Patologista Alocado','Exame']).mean()
display(faturamento_exame)
print("=+="*30)
print('Patologista - Exame - Ano - Espera')
periodo_exame = informacao_df[['Patologista Alocado','Ano','Exame', 'Espera']].groupby(['Patologista Alocado','Ano','Exame']).mean()
display(periodo_exame.head(40)) #head(40)) exibe 40 linhas
```
