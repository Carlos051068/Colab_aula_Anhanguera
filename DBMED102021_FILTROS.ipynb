```python
#https://www.youtube.com/watch?v=yV2BCFvXmC8
#Anhanguera
#Como importar arquivos .CSV para seu projeto Google Colab, Jupyter Notebook ou Spider

import pandas as pd

informacao_df = pd.read_csv('bdmed192021.csv') #importar arquivo bdmed192021.csv para COLAB

display(informacao_df.head(20))#mostra as 10 primeiras linhas
display(informacao_df.shape)#mostra quantidade de linha e colunas
display(informacao_df.describe())#mostra um resumo
print("="*90)
print('Exibe apenas a colunas selecionadas na lista []')
exame_patologista = informacao_df[['Espera','Tipo Exame','Exame','Patologista Alocado']]
display(exame_patologista)
print("="*90)
print('Exibe apenas a linha [2] ou [2:5] exibe até linha 5')
display(informacao_df.loc[2]) # Exibe(a série) a linha 2 do DataFrame [2:5]exibe até a linha 5
print("="*90) #pegar linhas sob uma condição

print('Seleciona o Patologista Alocado e seus exames linhas com condição.loc')
display(informacao_df.loc[informacao_df['Patologista Alocado'] == 'Dr. Eduardo Cambruzzi'])#pegar linhas sob uma condição
print("="*90)

print('Seleciona o Patologista Alocado e seus exames  várias linhas e colunas usando loc')
patol_df = informacao_df.loc[informacao_df['Patologista Alocado'] == 'Dra. Silvia Chaves e Silva',["Patologista Alocado","Espera", "Situação","Exame"]]
display(patol_df) #pegar várias linhas e colunas usando o loc
print("="*90)
#print(informacao_df.loc[1,'Usuario']) pegar 1 valor especifico na linha 1 da coluna Usuario = oliria
#informacao_df['Espera'].mean())
#CALCULAR INDICADORES
print('CALCULAR INDICADORES utilizando values_counts')
produtividade_pato = informacao_df['Patologista Alocado'].value_counts()
display(produtividade_pato)
print("="*90)
print('CALCULAR INDICADORES utilizando groupby mean média')
#sum = soma o total por exame ou mean média
faturamento_exame = informacao_df[['Exame','Espera']].groupby('Exame').mean()
display(faturamento_exame)
faturamento_blocos = informacao_df[['Exame','Quantidade de Blocos']].groupby('Exame').sum()
display(faturamento_blocos)
print('-'*90)
print('CALCULAR INDICADORES utilizando values_counts')
produtividade_exame = informacao_df['Exame'].value_counts()
display(produtividade_exame)
print("="*90)
```
