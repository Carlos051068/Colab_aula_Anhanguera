```python
import numpy as np
import pandas as pd
import plotly.express as px

vendas_df = pd.read_excel('vendas.xlsx')

print(vendas_df.head(6))#mostra as 10 primeiras linhas
display(vendas_df.shape)#mostra quantidade de linha e colunas
display(vendas_df.describe())#mostra um resumo
print('MOSTRA VALOR VAZIO *(DF_ISNA().ANY())*'*2)
display(vendas_df.isna().any()) #mostra valores vazios = True
print('|(DF_ISNA().ANY()) vazio=True*'*3)
print('-'*90)
print('*(DF_ISNA().SUM()) soma vazios*'*2)
display(vendas_df.isna().sum()) #SOMA valores vazios
print('|(DF_ISNA().SUM()) soma vazio*'*2)
print('=+='*19)
display(vendas_df)
print('=+='*19)
print(' GRÁFICO '*4)
gif = px.bar(vendas_df, x='Produto', y='Quantidade', color='ID Loja', barmode = 'group')
gif.show()
print(' GRÁFICO '*4)
```
