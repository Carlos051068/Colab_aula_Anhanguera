```python
#https://www.youtube.com/watch?v=L50fiScthSc
#Como manipular um DataFrame e EXPORTAR para CSV? | Análise de dados #9

import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import pandas as pd

informacao_df = pd.read_csv('bdmed192021.csv')

display(informacao_df.head(2))#mostra as 10 primeiras linhas
display(informacao_df.shape)#mostra quantidade de linha e colunas
display(informacao_df.describe())#mostra um resumo
informacao_df['Encerramento']= informacao_df['Encerramento'].fillna(method='ffill')#tratamento
informacao_df['Patologista Alocado']= informacao_df['Patologista Alocado'].fillna(method='ffill')#tratamento
print('TIPO DE DADO -'*3)
informacao_df.dtypes # exibe o tipo de cada coluna
print(informacao_df.dtypes)
print('TIPO DE DADO -'*3)
#print('MOSTRA VALOR VAZIO *(DF_ISNA().ANY())*'*2)
#display(informacao_df.isna().any()) #mostra valores vazios = True
#print('|(DF_ISNA().ANY()) vazio=True*'*3)
print('-'*90)
#print('*(DF_ISNA().SUM()) soma vazios*'*2)
#display(informacao_df.isna().sum()) #SOMA valores vazios
#print('|(DF_ISNA().SUM()) soma vazio*'*2)
print('=+='*19)

print("H"*90)
print('CALCULAR INDICADORES utilizando values_counts EXAME/ANO')
produtividade_ano = informacao_df[['Exame','Ano']].value_counts()
display(produtividade_ano)
produtividade_ano.to_csv('Produtividade_ano.csv')
print('produtividade_ano.to_csv("Produtividade_ano.csv")#criar csv do dataframe')
print('tipo de dado')
produtividade_ano.dtypes # exibe o tipo de cada coluna
print(produtividade_ano.dtypes)
print('tipo de dado')
print('CALCULAR INDICADORES utilizando values_counts Patologista Alocado')
print('CALCULAR INDICADORES utilizando values_counts')
produtividade_pato = informacao_df[['Patologista Alocado']].value_counts()
display(produtividade_pato)
produtividade_pato.to_csv('Produtividade_pato.csv')#criar csv do dataframe
print('produtividade_pato.to_csv("Produtividade_pato.csv")#criar csv do dataframe')
```
