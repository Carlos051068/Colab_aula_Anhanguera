```python
#https://www.youtube.com/watch?v=rVJW-rgJUkY&t=1339
#Anhanguera
#Como Tratar Valores Vazios em uma Base de Dados com Python
import pandas as pd
import numpy as np
import plotly.express as px

informacao_df = pd.read_csv('bdmed192021.csv') #importar arquivo bdmed192021.csv para COLAB

print('+ GRÁFICO'*5)

print('CALCULAR INDICADORES utilizando values_counts PRIORIDADE/QUANTIDADE DE BLOCOS')
prioridade_quantidade = informacao_df.groupby('Prioridade').sum()
prioridade_quantidade = prioridade_quantidade[['Quantidade de Blocos']].sort_values(by='Quantidade de Blocos', ascending=False)
display(prioridade_quantidade)

print('+ GRÁFICO'*5)

grafico = px.bar(prioridade_quantidade, x=prioridade_quantidade.index, y='Quantidade de Blocos',title = 'Prioridade/Quantidade de Blocos')
#gif = px.bar(x = dados_x, y = dados_y, title = 'Exame/Ano', height=400, width=900)

grafico.update_yaxes(title_font_color = 'blue')
grafico.update_yaxes(title_font_color = 'blue')
grafico.show()


#display(informacao_df.head(5))#mostra as 10 primeiras linhas
#print('*QUANTIDADE DE LINHAS*'*2)
#display(informacao_df.shape)#mostra quantidade de linha e colunas
#print('*RESUMO*'*10)
#display(informacao_df.describe())#mostra um resumo
#print('*RESUMO*'*10)
#print(''*90)
#print('*INFORMAÇÃO*'*5)
#display(informacao_df.info())#mostra informações dos dados
#print('+INFORMAÇÃO'*5)
#print('-'*90)
```
